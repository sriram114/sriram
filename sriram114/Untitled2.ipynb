```python
import requests
```

```python
url="https://www.flipkart.com/search?q=mobiles&otracker=AS_Query_HistoryAutoSuggest_2_0&otracker1=AS_Query_HistoryAutoSuggest_2_0&marketplace=FLIPKART&as-show=on&as=off&as-pos=2&as-type=HISTORY"
```

```python
responce=requests.get(url)
```

```python
responce
```

```python
soup=BeautifulSoup(responce.text)
```

```python
soup
```

```python
name=soup.select("._4rR01T")
```

```python
name
```

```python
price=soup.select("._1_WHN1")
```

```python
price
```

```python
import pandas as pd
```

```python
mobilelist=pd.DataFrame(name,columns=["mobilelist"])
mobilelist["price"]=price
```
